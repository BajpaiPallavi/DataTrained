# Solution to Q.1

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
soup = BeautifulSoup(page.content, 'html5lib')
table = soup.findAll('table')[1]
tableBody = table.find('tbody')
allVideos = []
#print(tableBody)
for row in table.findAll('tr')[1:-1]:
    video = {} 
    temp =row.findAll('td')[0].text
    temp = temp.split(".")
    video['Rank']= temp[0]
    Name =row.findAll('td')[1].text
    video['Name']= Name.split('"')[1]
    video['Artist']= row.findAll('td')[2].text
    video['Views']= row.findAll('td')[3].text
    video['Upload Date']= row.findAll('td')[4].text
    allVideos.append(video)
df = pd.DataFrame(allVideos)
df
```

# Solution to Q.7

```python
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome(r"/Users/pallavishu/Downloads/chromedriver")
driver.get("https://www.naukri.com/")
designation = driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys("Data science")
search = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[6]")
search.click()
jobs=[]
time.sleep(5)
mainDiv = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/section[2]/div[2]")
for row in mainDiv.find_elements(By.TAG_NAME,"article"):
    job = {}
    skillTemp = row.find_elements(By.TAG_NAME,'ul')[1]
    ulSkillTemp =skillTemp.text
    strUlSkillTemp = ulSkillTemp.split("\n")
    skills = ', '.join([str(elem) for elem in strUlSkillTemp])
    div =row.find_elements(By.TAG_NAME,'div')
    divDiv =div[0]
    divDivDiv =divDiv.find_element(By.TAG_NAME,'div')
    temp = divDivDiv.find_element(By.TAG_NAME,'div')
    
    aDivDivDiv= temp.find_element(By.TAG_NAME,'a')
    #print(aDivDivDiv.text)
    ulDivDiv = divDiv.find_element(By.TAG_NAME,'ul')
    liUlDivDiv = ulDivDiv.find_elements(By.TAG_NAME,'li')[2]
    aDivDiv = divDiv.find_element(By.TAG_NAME,'div')
    tTemp = aDivDiv.find_element(By.TAG_NAME,'a')
    job['Designation']=tTemp.text
    job['Location']=liUlDivDiv.text
    job['Company'] = aDivDivDiv.text
    job['Skills']= skills
    jobs.append(job)
df =pd.DataFrame(jobs)
df
```

# Solution to Q.8

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare#list')
soup = BeautifulSoup(page.content, 'html5lib')
tableBody = soup.find('tbody')
books =[]
for row in tableBody.findAll("tr"):
    book={}
    tds=row.findAll("td")
    book['Name']=tds[1].text
    book['Author']=tds[2].text
    book['Volume sold']=tds[3].text
    book['Publisher']=tds[4].text
    book['Genre']=tds[5].text
    books.append(book)
df = pd.DataFrame(books)
df
```

# Solution to Q.9

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get('https://www.imdb.com/list/ls095964455/')
soup = BeautifulSoup(page.content, 'html5lib')
mainDiv = soup.find('div',class_="lister-list")
allSeries=[]
for row in mainDiv.findAll('div',class_="lister-item mode-detail"):
    series={}
    divDetail =row.find('div',class_="lister-item-content")
    series['Name'] = divDetail.h3.a.text
    series['TS']=divDetail.h3.findAll('span')[1].text
    series['RunTime']=divDetail.p.findAll('span')[2].text
    temp=divDetail.div.find('div',class_="ipl-rating-star small")
    series['Rating'] = temp.findAll('span')[1].text
    temp=divDetail.findAll('p')[3]
    series['Votes'] = temp.findAll('span')[1].text
    allSeries.append(series)
df = pd.DataFrame(allSeries)
df
```
